#### Summary:
In a perfect world, a network would function perfectly, and all data on that
network would reliably make it to its final destination. Unfortunately, problems
occur frequently which cause data to arrive late, inconsistently, or not at all.
To fix this, protocols must be implemented to ensure reliable data transmission.
In this lab, we designed a reliable file transfer protocol which used UDP as a 
transport layer protocol which could handle packet drops, packet lag, and
duplicate packet sends. 

#### Design Choices:
Notable design choices include:
Designing the protocol around reliably transferring one packet at a time. 
This was implemented because it was more memory-friendly than the alternative 
method we came up with. The alternative involved storing packets in a bank
which, if a packet did not come in, the sender could back-track to the last 
message it knew it received. The flaw in this message, however, is depending on
the resources the computer has and the quality of the network, the bank may eat 
up all the memory of the computer. The method we implemented typically means
slower data transfer if network quality is bad, but more sends/receives will not
be taxing on the sender/receiver's resources.

Separation of responsibilities between the send file and send message. This 
design choice also was applied to the receive file/message. Because the FTP 
relies on reliably sending one packet before the next gets sent, all of the 
reliability functionality was implemented in the send/receive message. The 
responsibilities of the send/receive file methods were limited to anything file
related: send_file had to initiate a file send and pass information from the
file to the send_message method. receive_file had to write received data to a 
filepath, and detect when the end of the file was sent.

 2. Describe how your solution implements *tiering* and *layered* architecture.

The concept of layering is the logical separation of code to produce a 
coherent, but separated, set of funcitonality. A great example of layering in
this application is the logical separation of the send_file and send_message
methods. They each have their own functionality, which means the other can be 
changed, and as long as the inputs/outputs remain the same, they will still work
together. The concept of Tiering is the physical archetecture of the system that
is being created. In this case, there are two tiers: the client and the server.
